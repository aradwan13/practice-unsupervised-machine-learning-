Importing Modules 


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
sns.set()
```

### Loading data 

```python
df = pd.read_csv('mall-customers.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

### Data Extraction

```python
X = df.iloc[:,[3,4]].values
X
```

### Data Exploring

```python
plt.figure(figsize=(10,8))
sns.scatterplot(X[:,0],X[:,1])
```

Model buliding & Trainig 

### 1 Kmeans 

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
```

```python
kmeans.fit_transform(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
kmeans.labels_
```

```python
y_kmeans = kmeans.predict(X)
y_kmeans
```

```python
df_cluster = df.copy()
df_cluster['Cluster'] = y_kmeans
```

```python
df_cluster
```

### Model visulaization 

```python
plt.figure(figsize=(10,8))
sns.scatterplot(X[:,0],X[:,1],hue=kmeans.labels_,palette='gnuplot')
```

### Elbow Method 

```python
wcss = []
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)    
```

### 2 MiniBatchKMeans 

```python
from sklearn.cluster import MiniBatchKMeans
```

```python
mini = MiniBatchKMeans(n_clusters=5,random_state=42)
```

```python
mini.fit_transform(X)
```

```python
mini.cluster_centers_
```

```python
mini.inertia_
```

```python
mini.labels_
```

```python
y_mini=mini.predict(X)
y_mini
```

```python
df_cluster['Mini_cluster']=y_mini
```

```python
df_cluster
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(X[:,0],X[:,1],hue=mini.labels_,palette='rainbow')
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(X[:,0],X[:,1],hue=kmeans.labels_,palette='ocean')
```

Model Evaluation 

```python
mini.score(X)
```
